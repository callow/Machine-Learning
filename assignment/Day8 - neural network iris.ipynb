```python
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
```

```python
model  = Sequential() #顺序模型

#输入曾
model.add(Dense(4, input_shape=(4,))) # Dense 就是常用的全链接层次,里面有几个神经元
model.add(Activation("sigmoid")) # 激活函数

#隐层
model.add(Dense(160)) # Dense就是常用的全链接层
model.add(Activation("sigmoid")) #激活函数

#输出层
model.add(Dense(3)) # Dense就是常用的全链接层
model.add(Activation("softmax")) # 只能输出0 和 1

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
from sklearn.datasets import load_iris
from keras.utils import np_utils

iris = load_iris()

data_x = iris.data
data_y = iris.target

train_x = data_x
train_y = np_utils.to_categorical(data_y, 3) # 转换成one -hot-encoder

print (train_y[:3])

history = model.fit(train_x, train_y, epochs=500,  verbose=1) # verbose 打印结果  epoch 训练次数

print (history.history['acc'])
print (history.history['loss'])
```

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt


plt.xlabel('x')
plt.ylabel('y')
plt.title('acc graph')

plt.plot(range(len(history.history['acc'])), history.history['acc'])

plt.show()
```
